# prediksi model yang telah di training

```python
import cv2
import numpy as np
from keras.models import load_model
from matplotlib import pyplot as plt

model = load_model('model-007.model')
```

# Prediksi tidak pakai masker

```python
gambar = cv2.imread('C:/Users/Microsoft User/Documents/training/dat ujicoba/tidak pakai masker/21.jpg')
```

```python
plt.imshow(gambar)
```

```python
plt.imshow(cv2.cvtColor(gambar, cv2.COLOR_BGR2RGB))
```

```python
baca_gambar = cv2.resize(gambar, (100,100))
baca_gambar = np.expand_dims(baca_gambar, axis=0)
baca_gambar = baca_gambar/255.0
```

```python
#prediksi = model.predict(baca_gambar)
```

```python
prediksi =  model.predict(baca_gambar)

class_id = np.argmax(prediksi)

label_dict = ['tidak pakai masker', 'masker tidak sempurna', 'masker']

predicted_label = label_dict [class_id]
```

```python
print(predicted_label)
```

```python
confidence = prediksi[0][class_id]
```

```python
print('akurasi : %.2f'% (confidence*100))
```
